```python
# Importando a biblioteca pandas
import pandas as pd

# Lendo a carteira base do arquivo Excel
excel_file = "/kaggle/input/carteira-de-investimentos-fictcia/carteira_investimentos_ficticia.xlsx"
carteira_base = pd.read_excel(excel_file, sheet_name="Carteira Base")

# Inicializando uma lista para armazenar os resultados finais da exploração da carteira
carteira_explodida = []

# Iterando sobre cada linha da carteira base
for index, row in carteira_base.iterrows():
    
    # Lendo os valores relevantes para a análise de cada ativo
    nome_ativo_base = row["Asset's Name"][:31]  # Limitando o nome do ativo devido ao tamanho das abas
    valor_ativo_base = row["Asset's Value (thousand)"]
    participacao_base = row["Asset's Participation"]

    try:
        # Tentando acessar a aba específica do ativo no arquivo Excel
        aba_ativo_especifico = pd.read_excel(excel_file, sheet_name=nome_ativo_base)
        valor_total_carteira_base = 0  # Inicializando o valor total da carteira específica em uma variável
        dados_temporarios = []  # Lista temporária para armazenar os dados da aba específica para fazer o total da aba depois

        # Iterando sobre cada linha da aba específica do ativo
        for idx, row in aba_ativo_especifico.iterrows():
            nome_ativo_especifico = row["Asset's Name"]
            # Verificando se chegou ao final da aba específica (se chegou na linha de total do ativo ou se chegou em uma linha em branco)
            if nome_ativo_especifico == "Assets (R$ thousand)" or pd.isna(nome_ativo_especifico):
                break

            valor_ativo_especifico = row["Asset's Value (thousand)"]
            participacao_ativo_especifico = row["Asset's Participation"]
            # Calculando o valor explodido e a porcentagem explodida
            valor_explodido = valor_ativo_base * participacao_base * participacao_ativo_especifico
            percent_explodido = participacao_base * participacao_ativo_especifico
            # Atualizando o valor total da carteira base
            valor_total_carteira_base += valor_explodido
            # Adicionando os dados à lista temporária
            dados_temporarios.append([nome_ativo_base, nome_ativo_especifico, valor_explodido, percent_explodido])

        # Adicionando os dados da aba específica à lista principal com o valor total atualizado
        for dado in dados_temporarios:
            carteira_explodida.append(dado + [valor_total_carteira_base])

    except (ValueError, KeyError) as e:
        # Tratando o caso em que não há aba específica para o ativo
        valor_total_carteira = valor_ativo_base * participacao_base
        carteira_explodida.append([nome_ativo_base, "Não se aplica", valor_ativo_base, participacao_base, valor_total_carteira])

# Criando um DataFrame com os dados da carteira explodida
colunas = ["Carteira Base", "Nome do Ativo", "Valor Explodido", "% Explodida", "Total da Carteira"]
df_carteira_explodida = pd.DataFrame(carteira_explodida, columns=colunas)

# Exibindo o DataFrame resultante
df_carteira_explodida
```

```python
# Transformando o data frame em planilha excel

# Caminho para o arquivo Excel de saída
caminho_saida_excel = "/kaggle/working/carteira_explodida.xlsx"

# Salvar o DataFrame como uma planilha de Excel
df_carteira_explodida.to_excel(caminho_saida_excel, index=False)

print("Data Frame transformado em Excel com sucesso!")
```
